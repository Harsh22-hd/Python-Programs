```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['Maths','English','Science','Hindi']


plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y = ['Maths','English','Science','Hindi']

plt.pie(x, labels = y)
plt.legend()
plt.show()
```

```python
x = [10,20,30,40]
y = ['Maths','English','Science','Hindi']

c= ['blue','brown','black','yellow']

plt.pie(x, labels = y,colors = c)
plt.legend()
plt.show()
```
