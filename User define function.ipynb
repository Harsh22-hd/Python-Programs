```python
def calculate_exam(exam_scores):
    num_students = len(exam_scores)
    avg_score = sum(exam_scores) / num_students
    highest_score = max(exam_scores)
    lowest_score = min(exam_scores)

    num_passed = 0
    for score in exam_scores:
        if score >= 80:
            num_passed += 1

    num_failed = num_students - num_passed  # ✅ Correct formula

    return avg_score, highest_score, lowest_score, num_passed, num_failed

# --- Using the function ---
scores = [95, 80, 75, 100, 45, 88]

# Call the function and unpack the returned tuple into variables
avg, high, low, passed, failed = calculate_exam(scores)

print(f"Exam Results (for {len(scores)} students):")
print(f"  Average Score: {avg:.2f}")
print(f"  Highest Score: {high}")
print(f"  Lowest Score: {low}")
print(f"  Number Passed (>=80): {passed}")
print(f"  Number Failed (<80): {failed}")
```
