```python
num = 1
while num <= 50:
    if (num % 2 == 0):
        print(num)  # This line must be indented
    
    num = num + 1   # This line must be indented to be part of the 'while' loop
```

```python
email = input("Enter Your Email ID =")

valid_email = False

while not valid_email:
    if "@" in email and "." in email:
        print("Your Email address is valid")

        valid_email = True
    else:
        print("Your Email address is not valid...")
        email = input("Please enter valid Email =")
```

#for loop and if

```python
for i in range(0,10):
    if (i%2 == 0):
        print("The number is even",i)
    else:
        print("The number is odd",i)
```

```python
number = int(input("Enter the number = "))
print("The multiplication table of =",number)

for i in range(1,11):
    print(f"{number} x {i} = {number*i}")
```

```python
num_items = int(input("Enter the number of item ="))
total_price = 0

for i in range(num_items):
    price=float(input("Enter the price of item ="))
    total_price = total_price + price
    i = i + 1
if (total_price>=100):
    discount = 0.1*total_price
elif 50<=total_price<100:
    discount =0.5*total_price
else:
    discount = 0
    
total_bill = total_price - discount
print("Sub total =",total_bill)
```
