```python
def date_part(date_string):
    year = int(date_string[0:4])
    month = int(date_string[5:7])
    day = int(date_string[8:])
    return (year,month,day)
    
```

```python
date_part("2024-05-28")
```

```python
date_part("2025-06-22")
```
