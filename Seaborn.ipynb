```python
#Bar plot in Seaborn Library
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1['island']
```

```python
sns.barplot(x = df1.island , y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = 'island' , y = 'bill_length_mm' , data = df1 )
plt.show()
```

```python
sns.barplot(x = 'island' , y = 'bill_length_mm' , data = df1  ,hue = 'sex')
plt.show()
```

```python
#order parameter
```

```python
order1 = ['Dream' , 'Torgersen', 'Biscoe']
sns.barplot(x = 'island' , y = 'bill_length_mm',data =df1 , hue = 'sex', order = order1)
plt.show()
```

```python
#hue_order
```

```python
order1 = ['Dream' , 'Torgersen', 'Biscoe']
sns.barplot(x = 'island' , y = 'bill_length_mm',data =df1 , hue = 'sex', order = order1 , hue_order = ['Female','Male'])
plt.show()
```

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot( x = 'island' , y ='bill_length_mm' , data =df1 , color = 'pink')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue="sex" , palette = 'Accent')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue="sex" , palette = 'prism')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue="sex" , palette = 'gist_earth')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue="sex" , palette = 'gist_heat_r')
plt.show()
```

```python
#saturation parameter set contrast of color
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm' , data = df1 , hue="sex" , palette = 'gist_heat_r'
           ,saturation = 0.3)
plt.show()
```
