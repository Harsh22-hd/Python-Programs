```python
#set operations
```

```python
set1 = {1,2,3}
set2 = {3,4,5}

union_set = set1.union(set2)
print(union_set)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}

intersect_set = set1.intersection(set2)
print(intersect_set)
```

```python
set1 = {1,2,3}
set2 = {3,4,5}

Diff_set = set1.difference(set2)
Diff_set1 = set2.difference(set1)
print(Diff_set)
print(Diff_set1)
```

```python
#practice Questions
```

```python
student_passed_maths = {101,102,103,104,105}
student_passed_science = {103,104,105,106,107}

both_passed = student_passed_maths.intersection(student_passed_science)
print(both_passed)
print(len(both_passed))

only_math = student_passed_maths.difference(student_passed_science)
print(only_math)

only_science = student_passed_science.difference(student_passed_maths)
print(only_science)

sub = student_passed_maths.issubset(student_passed_science)
print(sub)

sub1 = student_passed_science.issuperset(student_passed_maths)
print(sub1)
```
