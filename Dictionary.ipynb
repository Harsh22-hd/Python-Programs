```python
squares = {x : x*x for x in range(6)}
print(squares)
```

```python
# Create an empty dictionary
products = {}

# Take input for at least 3 products
for i in range(3):
    product_id = int(input(f"Enter Product ID for product {i+1}: "))
    product_name = input(f"Enter Product Name for product {i+1}: ")
    
    # Store in dictionary
    products[product_id] = product_name

# Display the final dictionary
print("\nFinal Product Dictionary:")
print(products)
```
